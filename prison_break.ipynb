## Get the Data

Get the data from the [List of helicopter prison escapes](https://en.wikipedia.org/wiki/List_of_helicopter_prison_escapes) Wikipedia article.

```python
url = "https://en.wikipedia.org/wiki/List_of_helicopter_prison_escapes"
```

```python
data = data_from_url(url)
```

Print the first three rows.

```python
for row in data[:3]:
    print(row[:5])
```

```python
for row in data[:5]:
    date = row[0]
    year = fetch_year(date)
    print(year)
```

```python
# Extracting years and printing min/max years
years = [fetch_year(row[0]) for row in data]
min_year = min(years)
max_year = max(years)
print(f"Min year: {min_year}\nMax year: {max_year}")
```

```python
attempts_per_year = {}

# Loop through the dataset
for row in data:
    year = fetch_year(row[0])
    if year in attempts_per_year:
        attempts_per_year[year] += 1
    else:
        attempts_per_year[year] = 1

# Print the resulting dictionary
print(attempts_per_year)
```

```python
# Convert the dictionary to a list of [key, value] pairs
list_of_attempts = list(attempts_per_year.items())

# Call the barplot function with the correct input format
%matplotlib inline
barplot(list_of_attempts)
```
